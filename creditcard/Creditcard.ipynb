##### Outlier Detection and Handling:

```python
from scipy.stats import zscore

z_scores = zscore(df[['Amount']])
df_no_outliers = df[(z_scores < 3).all(axis=1)]
df_no_outliers
```

##### Split the data

```python
x = df.drop('Class',axis=1)
y = df['Class']
```

```python
x.shape
```

```python
y.shape
```

#### Training and Testing the data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

#### Training the Model

#### . Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)

# make predictions
y_pred1 = lr.predict(x_test)
```

```python
### Model Evaluation
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy1 = accuracy_score(y_test,y_pred1)
accuracy1
```

```python
confusion_mat = confusion_matrix(y_test,y_pred1)
sns.heatmap(confusion_mat,annot=True,cmap='viridis',fmt='.2f',linewidths=.5)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
### Cross validation
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(lr, x, y, cv=5, scoring='accuracy')
print("Cross-Validation Scores:", cv_scores)
cv_scores1 = np.mean(cv_scores)
print("Mean Accuracy:", cv_scores1)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=cv_scores)
plt.title('Cross-Validation Scores Distribution')
plt.xlabel('Cross-Validation Folds')
plt.ylabel('Accuracy')
plt.show()
```

##### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth=2, random_state=0)
rf.fit(x_train,y_train)

# Make predictions

y_pred2 = rf.predict(x_test)
y_pred2
```

```python
### Model Evaluation

accuracy2 = accuracy_score(y_test,y_pred2)
accuracy2
```

```python
conf_matrix = confusion_matrix(y_test, y_pred2)

# Visualize the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis', cbar=False,
            xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
print(classification_report(y_test,y_pred2))
```

```python
### Cross validation
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(rf, x, y, cv=5, scoring='accuracy')
print("Cross-Validation Scores:", cv_scores)
cv_scores2 = np.mean(cv_scores)
print("Mean Accuracy:", np.mean(cv_scores2))
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=cv_scores)
plt.title('Cross-Validation Scores Distribution')
plt.xlabel('Cross-Validation Folds')
plt.ylabel('Accuracy')
plt.show()
```

##### Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(x_train,y_train)

# Make predictions
y_pred3 = gnb.predict(x_test)
```

```python
### Model Evaluation

accuracy3 = accuracy_score(y_test,y_pred3)
accuracy3
```

```python
conf_matrix = confusion_matrix(y_test, y_pred3)

# Visualize the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis', cbar=False,
            xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.title('Naive Bayes Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=cv_scores)
plt.title('Cross-Validation Scores Distribution')
plt.xlabel('Cross-Validation Folds')
plt.ylabel('Accuracy')
plt.show()
```

##### DecisionTree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

# Make Predictions
y_pred4 = dt.predict(x_test)
```

```python
### Model Evaluation

accuracy4 = accuracy_score(y_test,y_pred4)
accuracy4
```

```python
conf_matrix = confusion_matrix(y_test, y_pred4)

# Visualize the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis', cbar=False,
            xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.title('Decision Tree Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plot_tree(dt, feature_names=df.columns, class_names=df.columns, filled=True)
plt.show()
```

```python
### Cross validation
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(dt, x, y, cv=5, scoring='accuracy')
print("Cross-Validation Scores:", cv_scores)
cv_scores4 = np.mean(cv_scores)
print("Mean Accuracy:", cv_scores4)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=cv_scores)
plt.title('Cross-Validation Scores Distribution')
plt.xlabel('Cross-Validation Folds')
plt.ylabel('Accuracy')
plt.show()
```

### Comparing the models

##### Before Cross-Validation:

Pros: Simplicity, quick evaluation.
Cons: Results may be sensitive to the particular random split, might not be representative of generalization.
##### After Cross-Validation:

Pros: More robust assessment of generalization, reduced risk of overfitting to a specific split.
Cons: Computationally more expensive, especially with a large number of folds.
